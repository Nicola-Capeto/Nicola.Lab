```python
Auto_file_3_df = Auto_file_3_df.loc[:,['Customer', 'ST', 'GENDER', 'Education', 'Customer Lifetime Value',
       'Income', 'Monthly Premium Auto', 'Number of Open Complaints',
       'Total Claim Amount', 'Policy Type', 'Vehicle Class']]
```

```python
customer_df=pd.concat([Auto_file_1_df,Auto_file_2_df,Auto_file_3_df], axis=0)
```

```python
customer_df.GENDER.unique()
```

```python
def clean_gender(x):
    if x in ['M', 'Male']:
        return 'Male'
    elif x in ['F', 'Femal','female']:
        return 'Female'
    elif np.nan:  pass
    else:
        return 'U'
```

```python
customer_df['GENDER'] = list(map(clean_gender, customer_df['GENDER'])) 
```

```python
customer_df
```

```python
customer_df=customer_df.drop_duplicates()
```

```python
customer_df
```

```python
customer_df.GENDER.unique()
```

```python
customer_df = customer_df.drop(labels=['Customer'], axis=1)
```

```python
customer_df
```

```python
customer_df["Customer Lifetime Value"] = customer_df["Customer Lifetime Value"].apply(lambda x: x.replace("%","") if type(x) == str else x)
```
