```python
def clean_gender(x):
    if x in ['M', 'Male']:
        return 'Male'
    elif x in ['F', 'Femal','female']:
        return 'Female'
    elif np.nan:  pass
    else:
        return 'U'
```

```python
customer_df['GENDER'] = list(map(clean_gender, customer_df['GENDER'])) 
```

```python
def clean_ST(x):
    if x in ['AZ']:
        return 'Arizona'
    elif x in ['WA']:
        return 'Washington'
    elif x in ['Cali']:
        return 'California'
    else:
        return x
```

```python
customer_df['ST'] = list(map(clean_ST, customer_df['ST'])) 
```

```python
def mean_Income(x):
    if x == 0 :
        return customer_df["Income"].mean()
    else:
        return x
```

```python
customer_df['Income'] = list(map(mean_Income, customer_df['Income'])) 
```

```python
def clean_Education(x):
    if x in ['Bachelors']:
        return 'Bachelor'
    else:
        return x
```

```python
def change_ST(x):
    if x in ['California']:
        return 'West Region'
    if x in ['Oregon']:
        return 'North West'
    if x in ['Washington']:
        return 'East'
    else:
        return 'Central'
```

```python
customer_df['ST'] = list(map(change_ST, customer_df['ST']))
```

```python
def Vehicle_Class(x):
    if x in ['Luxury SUV']:
        return 'Luxury Vehicle'
    elif x in ['Luxury Car']:
        return 'Luxury Vehicle'

    else:
        return x
```

```python
customer_df['Vehicle Class'] = list(map(Vehicle_Class, customer_df['Vehicle Class']))
```
